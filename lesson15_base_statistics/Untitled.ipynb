```python
# Выбросы
```

```python
Horsebase_columnscopy = Horsebase_columns.copy()
Horsebase_columnscopy = Horsebase_columnscopy.dropna(subset = {'pulse'})
```

```python
q1 = Horsebase_columnscopy['pulse'].quantile(0.25)
q3 = Horsebase_columnscopy['pulse'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = Horsebase_columnscopy[Horsebase_columnscopy['pulse'].between(lower_bound, upper_bound, inclusive=True)].sort_values('pulse')
pd.concat([Horsebase_columnscopy, remove_outliers]).drop_duplicates(keep=False)
```

Выбросы пульса высокие, они совпадают с высокой частотой дыхания, нормальня частота дыхания от 8 до 10, справочник говорит, что пульс у лошади и дыхание при беге или высокой температуре могут быть повышенные и достигать данных значений, значит выбросы нормальные.

```python
Horsebase_columnscopy = Horsebase_columns.copy()
Horsebase_columnscopy = Horsebase_columnscopy.dropna(subset = {'respiratory rate'})
```

```python
q1 = Horsebase_columnscopy['respiratory rate'].quantile(0.25)
q3 = Horsebase_columnscopy['respiratory rate'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = Horsebase_columnscopy[Horsebase_columnscopy['respiratory rate'].between(lower_bound, upper_bound, inclusive=True)].sort_values('respiratory rate')
pd.concat([Horsebase_columnscopy, remove_outliers]).drop_duplicates(keep=False)
```

Выбросы частоты дыхания присутствуют, одно значение 96 выбивается из всех показаний, но так как оно одно в группе высоких значений, которые могут быть в природе у лошадей, то мы его оставляем.

```python
Horsebase_columnscopy = Horsebase_columns.copy()
Horsebase_columnscopy = Horsebase_columnscopy.dropna(subset = {'rectal temperature'})
```

```python
q1 = Horsebase_columnscopy['rectal temperature'].quantile(0.25)
q3 = Horsebase_columnscopy['rectal temperature'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = Horsebase_columnscopy[Horsebase_columnscopy['rectal temperature'].between(lower_bound, upper_bound, inclusive=True)].sort_values('rectal temperature')
pd.concat([Horsebase_columnscopy, remove_outliers]).drop_duplicates(keep=False)
```

Выбросы температуры присутствуют, но они в пределах нормы.
