```python
–î–æ–º–∞—à–Ω–µ–µ –∑–∞–¥–∞–Ω–∏–µ "–ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑. –ü—Ä–æ–∏–∑–≤–æ–¥–Ω–∞—è".
–£—Ä–æ–≤–µ–Ω—å 0:
–ü–æ—Å—á–∏—Ç–∞–π—Ç–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–µ—Ä–≤–æ–π –∏ –≤—Ç–æ—Ä–æ–π –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏:
    
ùëì(ùë•)=ùë•5+4ùë†ùëñùëõ(2ùë•)+ùëêùëúùë†(3ùë•+3) 

–í —Ç–æ—á–∫–µ x = 1.

–£—Ä–æ–≤–µ–Ω—å 1:
–ü–æ—Å—Ç—Ä–æ–π—Ç–µ –≥—Ä–∞—Ñ–∏–∫ –ø–µ—Ä–≤–æ–π –∏ –≤—Ç–æ—Ä–æ–π –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏:
ùëì(ùë•)=ùë†ùëñùëõ5(2ùë•+1) 
–ù–∞ –æ—Ç—Ä–µ–∑–∫–µ [-5; 5].
```

```python
import numpy as np
import math
from scipy.misc import derivative
import matplotlib.pyplot as plt
```

–ü–æ—Å—á–∏—Ç–∞–π—Ç–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–µ—Ä–≤–æ–π –∏ –≤—Ç–æ—Ä–æ–π –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏:

```python
def f(x):
    return x**5 + 4*math.sin(2*x) + math.cos(3*x+3)
```

```python
derivative(f, 1.0 , dx=1e-6, n=1 ) #–ø–µ—Ä–≤–∞—è –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–∞—è –≤ —Ç–æ—á–∫–µ 1
```

```python
derivative(f, 1.0 , dx=1e-6, n=2 ) #–≤—Ç–æ—Ä–∞—è –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–∞—è –≤ —Ç–æ—á–∫–µ 1 
```

–£—Ä–æ–≤–µ–Ω—å 1:
–ü–æ—Å—Ç—Ä–æ–π—Ç–µ –≥—Ä–∞—Ñ–∏–∫ –ø–µ—Ä–≤–æ–π –∏ –≤—Ç–æ—Ä–æ–π –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏:
ùëì(ùë•)=ùë†ùëñùëõ5(2ùë•+1) 
–ù–∞ –æ—Ç—Ä–µ–∑–∫–µ [-5; 5].

```python
def f1(x):
    return math.sin(2*x+1)**5
```

```python
x = np.linspace(-5, 5)
```

```python
der1 = [derivative(f1, xx, dx=1e-6, n=1) for xx in x]
der2 = [derivative(f1, xx, dx=1e-6, n=2) for xx in x]
```

```python
plt.plot(x, der1, label="derivative1")
plt.legend()
plt.show()
```

```python
plt.plot(x, der2, label="derivative2")
plt.legend()
plt.show()
```
