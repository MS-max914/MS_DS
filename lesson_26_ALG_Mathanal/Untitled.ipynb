```python
import numpy as np
import math
from scipy.misc import derivative
import matplotlib.pyplot as plt
```

Посчитайте значение первой и второй производной функции:

```python
def f(x):
    return x**5 + 4*math.sin(2*x) + math.cos(3*x+3)
```

```python
derivative(f, 1.0 , dx=1e-6, n=1 ) #первая производная в точке 1
```

```python
derivative(f, 1.0 , dx=1e-6, n=2 ) #вторая производная в точке 1 
```

Уровень 1:
Постройте график первой и второй производной функции:
𝑓(𝑥)=𝑠𝑖𝑛5(2𝑥+1) 
На отрезке [-5; 5].

```python
def f1(x):
    return math.sin(2*x+1)**5
```

```python
x = np.linspace(-5, 5)
```

```python
der1 = [derivative(f1, xx, dx=1e-6, n=1) for xx in x]
der2 = [derivative(f1, xx, dx=1e-6, n=2) for xx in x]
```

```python
plt.plot(x, der1, label="derivative1")
plt.legend()
plt.show()
```

```python
plt.plot(x, der2, label="derivative2")
plt.legend()
plt.show()
```
