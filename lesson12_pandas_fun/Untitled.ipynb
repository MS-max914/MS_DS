```python
# Задание 1**
# Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
# - оценка 2 и меньше - низкий рейтинг
# - оценка 4 и меньше - средний рейтинг
# - оценка 4.5 и 5 - высокий рейтинг

# Результат классификации запишите в столбец class
```

```python
import pandas as pd
```

```python
rating = pd.read_csv('ratings.csv')
rating.head()
```

```python
def clasif (rating_):
    if rating_<=2:
       return 'low_rating'
    if rating_ <=4:
       return 'medium_rating'
    if rating_ >=4.5:
       return 'hight rating'
    
rating = pd.read_csv('ratings.csv')
rating['class'] = rating.rating.apply(clasif)   
rating.head(10)
```

```python
# Задание 2
# Используем файл keywords.csv.

# Необходимо написать гео-классификатор, который каждой строке сможет выставить
# географическую принадлежность определенному региону. Т. е. если поисковый запрос
# содержит название города региона, то в столбце ‘region’ пишется название этого региона. 
# Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

# Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

# geo_data = {


# 'Центр': ['москва', 'тула', 'ярославль'],

# 'Северо-Запад': ['петербург', 'псков', 'мурманск'],

# 'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
# }

# Результат классификации запишите в отдельный столбец region.
```

```python
keywords = pd.read_csv('keywords.csv')

geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

keywords.head()
```

```python
def region_search(geo):                           
    for region in geo_data:                       
        for city in geo_data[region]:             
            if city in geo:                       
                return region
    return 'undefined'

keywords['region'] = keywords['keyword'].apply(region_search)  
keywords.region.value_counts()                      
```
